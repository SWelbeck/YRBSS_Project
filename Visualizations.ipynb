# Visualizations

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["figure.figsize"] = [10,5]
```

```python
df = pd.read_csv('final_df.csv')
```

**Descriptive Statistics**

```python
df.describe()
```

**Stacked bar-chart for condom use by age group**

From the visualization we can see that she share of respondants who used a condom that last time that had sexual intercourse increases with age. This share is the most wide at 17 years old.

```python
# Passenger class summary
print('Condom Usage by Age Group: \nThe last time you had sexual intercourse, did you or your partner use a condom?')

print('\n')
print(df.groupby(['How old are you','Condom use'])['How old are you'].count().unstack())

# Passenger class visualization
age = df.groupby(['How old are you','Condom use'])['How old are you'].count().unstack()
p1 = age.plot(kind = 'bar', stacked = True, 
                   title = 'Condom Usage by Age Group', 
                color = ['red','pink','grey'], alpha = .70)
x = [0,1,2,3,4,5,6]
labels = ['12 or younger','13','14','15','16','17','18 or older']
p1.set_xlabel('Age')
p1.set_ylabel('# of respondants')
plt.xticks(x, labels,rotation = 45)

p1.legend(['Never had sex','Did use condom','Did not use condom'])

plt.show()
```

**Condom Use Over Time**

This visualization shows us that over time the share of respondants who did use a condom has become narrower. 

```python
# Passenger class summary
print('Condom Use Over Time: \nThe last time you had sexual intercourse, did you or your partner use a condom?')

print('\n')
print(df.groupby(['YEAR','Condom use'])['YEAR'].count().unstack())

# Passenger class visualization
age = df.groupby(['YEAR','Condom use'])['YEAR'].count().unstack()
p1 = age.plot(kind = 'bar', stacked = True, 
                   title = 'Condom Use Over Time', 
                   color = ['red','pink','grey'], alpha = .70)
p1.set_xlabel('YEAR')
plt.xticks(rotation = 45)
p1.set_ylabel('# of respondants')
p1.legend(['Never had sex','Did use condom','Did not use condom'])
plt.show()
```

**Age Groups Under the Influence During Sex**

The graph below shows us the those who have had sex under the influence per age group is relatively small.

```python
# Passenger class summary
print('Alcohol/drugs and sex: \nDid you drink alcohol or use drugs before you had sexual intercourse the last time?')

print('\n')
print(df.groupby(['How old are you','Alcohol/drugs and sex'])['How old are you'].count().unstack())

# Passenger class visualization
age = df.groupby(['How old are you','Alcohol/drugs and sex'])['How old are you'].count().unstack()
p1 = age.plot(kind = 'bar', stacked = True, 
                   title = 'Age Groups Under the Influence During Sex', 
                   color = ['red','pink','grey'], alpha = .70)
x = [0,1,2,3,4,5,6]
labels = ['12 or younger','13','14','15','16','17','18 or older']
plt.xticks(x, labels,rotation = 45)
p1.set_xlabel('Age')
p1.set_ylabel('# of Respondants')
p1.legend(['Never had sex','Was under the influence','Was not under the influence'])
plt.show()
```

```python
# Passenger class summary
print('Alcohol/drugs and sex')

print('\n')
print(df.groupby(['How old are you','Ever sexual intercourse'])['How old are you'].count().unstack())

# Passenger class visualization
age = df.groupby(['How old are you','Ever sexual intercourse'])['How old are you'].count().unstack()
p1 = age.plot(kind = 'bar', stacked = True, 
                   title = 'Condom Use by Age', 
                   color = ['red','pink','grey'], alpha = .70)
p1.set_xlabel('Age')
x = [0,1,2,3,4,5,6]
labels = ['12 or younger','13','14','15','16','17','18 or older']
plt.xticks(x, labels,rotation = 45)
p1.set_ylabel('# of respondants')
p1.legend(['Yes','No'])
plt.show()
```

```python
# Count of 'Considered Suicide'
print('Target Vari')
print(df.groupby(['Considered suicide'])['Considered suicide'].count())

# Target Variable Countplot
sns.set_style('darkgrid')
plt.figure(figsize = (10,5))
sns.countplot(df['Considered suicide'], alpha =.80, palette= ['grey','lightgreen'])
plt.title('Considered Suicide')
plt.ylabel('# Students')
plt.show()
```

```python
# Count of 'Considered Suicide'
print('Target Variable')
print(df.groupby(['Attempted suicide'])['Attempted suicide'].count())

# Target Variable Countplot
sns.set_style('darkgrid')
plt.figure(figsize = (10,5))
sns.countplot(df['Attempted suicide'], alpha =.80, palette= ['grey','lightgreen'])
plt.title('Attempted Suicide')
plt.ylabel('# Students')
plt.show()
```
